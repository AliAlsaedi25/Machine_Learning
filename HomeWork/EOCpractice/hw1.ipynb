```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
```

```python
data_directory = "C:\\Users\\alsae\\Desktop\\recmond\\ara218\\HW1\\data\\"

bikes = pd.read_csv(f"{data_directory}day.csv")

bikes
```

```python
first_rows = bikes.head()
first_rows
```

```python
data_info = bikes.info()
data_info
```

```python
data_stats = bikes.describe()
data_stats
```

```python
mean_value = data_stats["temp"]["mean"]
filtered_dataframe = bikes[bikes['temp'] > mean_value]
row_count = len(filtered_dataframe)
#row_count = filtered_dataframe.count()
print(row_count)

```

```python
num_array = bikes.to_numpy()
print(num_array.shape)
print(num_array[99:105])
```

```python
new_array = num_array[:,9:13]
sorted_array = new_array[new_array[:,1].argsort()]
print(sorted_array[:5])
```

```python
mean_riders_per_day = bikes.groupby("weekday")["cnt"].mean().reset_index()

mean_riders = sns.catplot(x="weekday", y="cnt", data=mean_riders_per_day)

```

```python
codes = pd.CategoricalDtype(categories=['sunday','monday','tuesday','wednesday','thursday','friday','saturday'],ordered=True)
bikes['day_names'] = pd.Categorical.from_codes(bikes['weekday'], dtype=codes)
bikes.head()
```

```python
mean_riders_per_day = bikes.groupby("day_names")["cnt"].mean().reset_index()

mean_riders = sns.catplot(x="day_names", y="cnt", data=mean_riders_per_day)
```

```python
line_plot = sns.lineplot(data=bikes, x=bikes["instant"], y=bikes["cnt"])
line_plot
```

```python
bikes['dt'] = pd.to_datetime(bikes['dteday'])
_plot = sns.lineplot(data=bikes, x=bikes["dt"], y=bikes["cnt"])
_plot
```
